```python
df_train_test= df_train.copy()

def corr_i(i):
    test = df_train_test.corr(numeric_only = True).unstack().to_frame().reset_index().dropna()
    test.rename(columns ={0 :"corr"} ,inplace =True)
    test = test.loc[test['corr']!=1]
    test = test.loc[test.level_0 == 'label'].sort_values('corr',ascending = False).drop_duplicates('corr')
    test = test.iloc[:,1:].reset_index(drop = True)
    test=test.loc[test.level_1 ==i]
    return test
corr_i(i)
# test.loc[:,'corr'].values
# 독립변수들의 범주화를 위해 최적의 상관관계를 찾기 위한 1번쨰 함수
```

```python
i = 'BMI'
d = 1
i_min = df_train[i].min()
i_max = df_train[i].max()

std_t = [x for x in np.arange(i_min, i_max, d)]
df_col_corr = pd.DataFrame(std_t, columns = ['변수'])
df_col_corr['값'] = 0
t = 0

for j in std_t:

    def m_p(x):
        if x > j :
            return 1
        else : 
            return 0

    df_train_test[i] = df_train[i].apply(m_p)
#     print(df_train_test.mean())

    corr_i(i)
    
    try:
        df_col_corr.loc[t,'값'] = corr_i(i).loc[:,'corr'].values    
    except:
        break
    t = t+1

df_col_corr.loc[df_col_corr['값']!=0].sort_values('값',ascending = False).reset_index(drop = True)
```

```python
#데이터 쪼개기 
# Y = df_train.label
df_train.drop(columns = ['ID'], inplace = True)
X_train = df_train.copy()
# X_train_ID = df_train.ID
df_test = df_test.iloc[:,1:]
```

```python
X_train.drop(columns = ['요 단백','충치'],inplace = True)
df_test.drop(columns = ['요 단백','충치'],inplace = True)
```

```python
smote = SMOTE(random_state=42)
X_train_over, y_train_over = smote.fit_resample(X_train, Y)
#오버 샘플링
```

```python
print(X_train_over.shape)
print(y_train_over.shape)
```

```python

# 스케일러
from sklearn.preprocessing import StandardScaler
# scaler =['나이', '키(cm)', '몸무게(kg)', 'BMI', '시력', '충치', '공복 혈당', '혈압', '중성 지방',
#        '혈청 크레아티닌', '콜레스테롤', '고밀도지단백', '저밀도지단백', '헤모글로빈', '요 단백', '간 효소율']
scaler =['헤모글로빈','키(cm)','중성 지방','간 효소율','고밀도지단백', '공복 혈당','시력','나이',  'BMI']

# scaler =[ '헤모글로빈', '몸무게(kg)','중성 지방','고밀도지단백','나이','공복 혈당']

st = StandardScaler()

st.fit(X_train_over[scaler])

X_train_over[scaler] = st.transform(X_train_over[scaler])
df_test[scaler] = st.transform(df_test[scaler])

from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(X_train_over, y_train_over,test_size = 0.3, random_state = 42,stratify =y_train_over)


from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 42)
rf.fit(x_train, y_train)

predict_train = rf.predict(x_train)
predict_val = rf.predict(x_val)

from sklearn.metrics import accuracy_score ,f1_score, confusion_matrix, classification_report
print("\n>>>train --------------------------------")
print("accuracy_score : ", accuracy_score(y_train, predict_train))
# print("f1_score : ", f1_score(y_train, predict_train))

print("\n>>>test ---------------------------------")
print("accuracy_score : ", accuracy_score(y_val, predict_val))
# print("f1_score : ", f1_score(y_val, predict_val))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':[400,450,500],'max_depth' : [16,17,18], 'min_samples_leaf':[1,2,3]}
clf = GridSearchCV(rf, param_grid, cv = 3)
clf.fit(x_train, y_train)
print('Best Parameters: ', clf.best_params_)
print('Best Score: ', clf.best_score_)
print('TestSet Score: ', clf.score(x_val, y_val))
```
